```python

import pandas as pd
from CURLS.preprocess import DataPreprocessor
from CURLS.rule_learning import CausalRuleLearner
```

```python
# 数据预处理
synthetic_data_path = "../data/sys_data2.csv"
# 协变量的列
covariate_columns = ["X1", "X2", "X3", "X4", "X5", "X6", "X7", "X8", "X9", "X10"]
# 类别型变量的列
col_categ = ["X1", "X2"]
df = pd.read_csv(synthetic_data_path)
data_preprocessor = DataPreprocessor(covariate_columns, col_categ, num_thresh=8)
df_binarized = data_preprocessor.fit_transform(df)
```

```python
# 规则学习
causal_learner = CausalRuleLearner(
    df_binarized, max_rule_length=4, variance_weight=0.2, minimum_coverage=0.08
)
causal_learner.find_rules(num_rules_to_find=2)
causal_learner.print_rules()
```
